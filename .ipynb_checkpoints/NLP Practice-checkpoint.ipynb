#### to compare pairs of words to see which pair is "more likely to occur

```python
import nltk.collocations
import nltk.corpus
import collections
nltk.download('brown')

bgm    = nltk.collocations.BigramAssocMeasures()
finder = nltk.collocations.BigramCollocationFinder.from_words(
    nltk.corpus.brown.words())
scored = finder.score_ngrams( bgm.likelihood_ratio  )

# Group bigrams by first word in bigram.                                        
prefix_keys = collections.defaultdict(list)
for key, scores in scored:
   prefix_keys[key[0]].append((key[1], scores))

# Sort keyed bigrams by strongest association.                                  
for key in prefix_keys:
   prefix_keys[key].sort(key = lambda x: -x[1])

print ('car', prefix_keys['car'][:20])
```

##### Text diversity score function

```python
# function to calculate a diversity score
def lexical_diversity(text):
    word_count = len(text)
    vocab_size = len(set(text))
    diversity_score = word_count / vocab_size
    return diversity_score
```

```python
print(text,round(lexical_diversity(text),2))
print(10 * '%')
text1 = 'ciao io sono giorgio'
print(text1,round(lexical_diversity(text1),2))
```

##### Function for plural

```python
def plural(word):
    if word.endswith('y'):
        return word[:-1] + 'ies'
    elif word[-1] in 'sx' or word[-2:] in ['sh','ch']:
        return word + 'es'
    elif word.endswith('an'):
        return word[:-2] + 'en'
    else:
        return word + 's'
# notice that endswith is called 'method' because it is a function associateed to a string object
```

```python
print('man',plural('man'))
print('fly', plural('fly'))
print('match', plural('match'))

print('fan',plural('fan'), '--> wrong. It should be: fans')
```

##### unusual_words :  misspelled words - uncommon words

```python
def uncommon_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words()) # english vocabulary
    unusual = text_vocab.difference(english_vocab) # difference function
    return sorted(unusual)
```

##### function to compute what fraction of words in a text are not in the stopwords list

```python
def content_fraction(text):
    list_stopwords = stopwords.words('english')
    content = [w for w in word_tokenize(text) if w.lower() not in list_stopwords]
    return round(len(content)/len(text),3)
```

```python
print('text is :', text)
print('the fraction of non stop words is:', content_fraction(text))
print(20*'%')
print('words in text that are stopwords:', [w for w in word_tokenize(text) if w in stopwords.words('english')])
```

From previous analysis cell 495, we saw that male and female names have some distinctive characteristcs:
names ending in a,e, and i are likely to be female while names ending in k,o,r,s, and t are likely to be male.

We can build a classifier to model these differences more precisely.

The first step in creating a classifier is deciding what features of the input are relevant and how to encode those features. 

```python
def gender_feature(word):
    return {'last_letter': word[-1]}

gender_feature('Shrek')
```

```python
from nltk.corpus import names
nltk.download('names')
import random
names = ([(name,'male') for name in names.words('male.txt')] + 
         [(name, 'female') for name in names.words('female.txt')])
random.shuffle(names)
```

```python
names
```

```python
### we are going to train a Naive Bayes
featuresets = [(gender_feature(n),g) for (n,g) in names]
featuresets
```

```python
train_set, test_set = featuresets[500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set) # nltk bayes
```

```python
# example
print('Gender prediction for NEO:',classifier.classify(gender_feature('Neo')))
print('Gender prediction for Trinity:',classifier.classify(gender_feature('Trinity')))
print(40*'%')
print('evaluation on testing set')
print(nltk.classify.accuracy(classifier, test_set)) # nltk accuracy
```

we can examine the classifier to determine which features it found most effective for distinguishing the names'genders

#### Theory:
Typically, feature extractors are built through a process of trial and error, guided by intuitions about what information is relevant to the problem. It is common to start with a 'kitchen sink' approach, including all the features that we can think of, and then checking to see which features actually are helpful.

###### example of overfitting when creating features for training
the featuresets we are going to create contain a large number of specific features, leading to overfitting for the relatively small Names Corpus

```python
def gender_features2(name):
    ''' dictionary with letter count in a word '''
    features = {}
    features['firstletter'] = name[0].lower()
    features['lastletter'] = name[-1].lower()
    for letter in 'abcdefghijklmnopqrstuvwxyz':
        features['count(%s)' % letter] = name.lower().count(letter)
        features['has(%s)' %letter] = (letter in name.lower())
    return features
```

```python
print(gender_features2('Johnson'))
```

##### Theory:
We should not provide so many features because otherwise the algorithm will have a higher chance of relying on specific information/characteristics (idiosyncrasies) of the training data; thereby we fail to generalize well to new examples

```python
featuresets = [(gender_features2(n),g) for (n,g) in names]
train_set, test_set = featuresets[500:], featuresets[:500] # featuresets is a list
classifier = nltk.NaiveBayesClassifier.train(train_set) # nltk bayes
# example
print('Gender prediction for NEO:',classifier.classify(gender_features2('Neo')))
print('Gender prediction for Trinity:',classifier.classify(gender_features2('Trinity')))
print(40*'%')
print('evaluation on testing set')
print(nltk.classify.accuracy(classifier, test_set)) # nltk accuracy
```

##### Theory
Once an initial set of features has been chosen, a productive method for refining the feature set is error analysis:
First, we select a development set, containing the corpus data for creating the model. 
This development set is then subdivided into the trainingset and the dev-test set

```python
train_names = names[1500:]
devtest_names = names[500:1500] # the devtest is used for error analysis. 
test_names = names[:500]
```

```python
train_set = [(gender_feature(n),g) for (n,g) in train_names]
devtest_set = [(gender_feature(n),g) for (n,g) in devtest_names]
test_set = [(gender_feature(n),g) for (n,g) in test_names]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

using the dev-test set, we can generate a list of the errors that the classifier makes when predicting name genders:
    

```python
# list of errors that the classifier makes
errors = []
for (name,tag) in devtest_names: # tag would be male or female
    guess = classifier.classify(gender_feature(name))
    if guess != tag:
        errors.append((tag,guess,name))
```

```python
len(errors)/len(devtest_names)
```

```python
#!pip install pandas
import pandas as pd
pd.DataFrame(errors, columns=  ['actual label','predicted','name'])
```

we can then examine individual error cases where the model predicted the wrong label, and try to determine what additional pieces of information would allow it to maek the right decision or which existing pieces of information are tricking it into making the wrong decision. 

```python
for (tag,guess,name) in sorted(errors):
    print('correct:%-8s guess=%-8s name = %-30s' % (tag,guess,name))
```

If we have a look at the list of errors makes it clear that some suffixes that are more than one letter can be indicative of name genders. For example, names ending in 'yn' appear to be predominantly female, despite the fact that names ending in 'n' tend to be male and names ending in 'ch' are usually male.

```python
def gender_Features(word):
    return {'suffix1':word[-1:],
           'suffix2':word[-2:]}
```

we rebuild the classifier with the mew feature ectractor, we see that the performance on the dev-test went up.

```python
train_set = [(gender_feature(n),g) for (n,g) in train_names]
devtest_set = [(gender_feature(n),g) for (n,g) in devtest_names]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier, devtest_set))
```

#### Theory
The error analysis procedure can be repeated checking for patterns in the errors that are made by the newly improved classifier. Each time the error analysis procedure is repeated, we should select a different dev-test/training split to ensure that the classifier does not start to reflect idiosyncrasies in the dev-test set. 
However, once we have used the dev-test set to help us the develop the model, we can no longer trust that it will give us an accurate idea of how well the model would perform on new data. It is therefore important to keep the test set separate and unused until our model development is complete.
